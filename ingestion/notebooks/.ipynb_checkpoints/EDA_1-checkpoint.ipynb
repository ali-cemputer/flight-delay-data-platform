## Bölüm 2 — Null / Eksik Veri Analizi
Null'ların nerede ve neden oluştuğunu anlamak, pipeline'da doğru strateji kurmak için kritik.

### 2.1 Null Sayısı ve Oranı

```python
null_counts = df.isnull().sum()
null_pct    = (null_counts / len(df) * 100).round(2)

null_df = (
    pd.DataFrame({'null_count': null_counts, 'null_pct': null_pct})
    .query('null_count > 0')
    .sort_values('null_pct', ascending=False)
)

print(f"Null içeren kolon sayısı: {len(null_df)} / {len(df.columns)}\n")
print(null_df.to_string())
```

### 2.2 Isı Haritası — Null Dağılımı
_İlk 10.000 satır üzerinde — büyük veri için yeterli temsil._

```python
cols_with_nulls = null_df.index.tolist()
sample_df       = df.sample(10_000, random_state=42)

fig, ax = plt.subplots(figsize=(14, 6))
sns.heatmap(
    sample_df[cols_with_nulls].isnull(),
    yticklabels=False,
    cbar=False,
    cmap='viridis',
    ax=ax
)
ax.set_title('Null Dağılımı (10k satır sample) — Sarı = Null', fontsize=13)
ax.set_xlabel('Kolonlar')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### 2.3 Null Grupları — Beklenen mi, Beklenmedik mi?

```python
# Delay columns should be ~100% null for cancelled flights
delay_cols    = ['CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay']
cancelled     = df[df['Cancelled'] == 1]
not_cancelled = df[df['Cancelled'] == 0]

print("--- Delay kolonları — iptal uçuşlarda null oranı (beklenen: ~%100) ---")
print((cancelled[delay_cols].isnull().mean() * 100).round(2))

print("\n--- Delay kolonları — gerçekleşen uçuşlarda null oranı (beklenen: düşük) ---")
print((not_cancelled[delay_cols].isnull().mean() * 100).round(2))
```

```python
# CancellationCode: populated for cancelled, null for the rest
code_in_cancelled     = df[df['Cancelled'] == 1]['CancellationCode'].isnull().mean() * 100
code_in_not_cancelled = df[df['Cancelled'] == 0]['CancellationCode'].isnull().mean() * 100

print("CancellationCode null oranı:")
print(f"  İptal edilenlerde   : {code_in_cancelled:.1f}%  (beklenen: ~0%)")
print(f"  İptal edilmeyenlerde: {code_in_not_cancelled:.1f}%  (beklenen: ~100%)")
```

```python
# DepTime / ArrTime: should be null for cancelled flights
time_null = df[df['Cancelled'] == 1][['DepTime', 'ArrTime']].isnull().mean() * 100

print("DepTime / ArrTime — iptal uçuşlarda null oranı (beklenen: ~%100):")
print(time_null.round(1))
```

### 2.4 Kritik Null'lar — Pipeline İçin İşaretlenenler

```python
critical_nulls = (
    not_cancelled.isnull().mean() * 100
).round(2)

critical_nulls = critical_nulls[critical_nulls > 5].sort_values(ascending=False)

print("Gerçekleşen uçuşlarda >%5 null olan kolonlar (pipeline riski):")
print(critical_nulls.to_string())
```

## Bölüm 5 — Gecikme Analizi

```python
# is_delayed definition used in the pipeline
delayed_rate = (df['ArrDelay'] > 15).mean() * 100
print(f"ArrDelay > 15 dk olan uçuş oranı (is_delayed): {delayed_rate:.2f}%")
```

```python
# Clip to -60/+180 so extreme outliers don't squash the histogram
arr_delay = df['ArrDelay'].dropna()
neg_pct   = (arr_delay < 0).mean() * 100

fig, ax = plt.subplots(figsize=(10, 4))
arr_delay.clip(-60, 180).hist(bins=80, ax=ax, edgecolor='none')
ax.axvline(0,  color='red',    linestyle='--', linewidth=1.2, label='0 dk')
ax.axvline(15, color='orange', linestyle='--', linewidth=1.2, label='15 dk eşiği')
ax.set_title('ArrDelay Dağılımı (−60 / +180 dk arası kırpılmış)', fontsize=13)
ax.set_xlabel('Varış Gecikmesi (dk)')
ax.set_ylabel('Uçuş Sayısı')
ax.legend()
plt.tight_layout()
plt.show()

print(f"Negatif ArrDelay oranı (erken iniş): {neg_pct:.1f}%")
```

```python
def categorize_delay(x):
    if pd.isna(x):  return 'Unknown'
    elif x <= 0:    return 'No Delay'
    elif x <= 15:   return 'Minor (1-15 dk)'
    elif x <= 60:   return 'Major (16-60 dk)'
    else:           return 'Severe (60+ dk)'

cat_order  = ['No Delay', 'Minor (1-15 dk)', 'Major (16-60 dk)', 'Severe (60+ dk)', 'Unknown']
delay_cats = df['ArrDelay'].apply(categorize_delay).value_counts().reindex(cat_order, fill_value=0)
delay_pct  = (delay_cats / len(df) * 100).round(2)

print(pd.DataFrame({'count': delay_cats, 'pct': delay_pct}).to_string())
```

```python
# Total delay minutes contributed by each cause
delay_cols   = ['CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay']
cause_totals = df[delay_cols].sum()
cause_pct    = (cause_totals / cause_totals.sum() * 100).round(2)

fig, ax = plt.subplots(figsize=(8, 4))
cause_pct.plot(kind='bar', ax=ax, edgecolor='white')
ax.set_title('Gecikme Nedenlerinin Payı (Toplam Gecikme Dakikası %)', fontsize=13)
ax.set_xlabel('Gecikme Nedeni')
ax.set_ylabel('%')
ax.tick_params(axis='x', rotation=20)
plt.tight_layout()
plt.show()
```

```python
corr_cols = ['CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay', 'ArrDelay']
corr      = df[corr_cols].corr()

fig, ax = plt.subplots(figsize=(7, 5))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', center=0, ax=ax)
ax.set_title('Delay Kolonları Korelasyon Matrisi', fontsize=13)
plt.tight_layout()
plt.show()
```

```python
# Sample 10k — full dataset makes scatter unreadable
sample = df[['DepDelay', 'ArrDelay']].dropna().sample(10_000, random_state=42)

fig, ax = plt.subplots(figsize=(7, 5))
ax.scatter(sample['DepDelay'], sample['ArrDelay'], alpha=0.15, s=5)
ax.axhline(0, color='red', linestyle='--', linewidth=0.8)
ax.axvline(0, color='red', linestyle='--', linewidth=0.8)
ax.set_title('DepDelay vs ArrDelay (10k sample)', fontsize=13)
ax.set_xlabel('Kalkış Gecikmesi (dk)')
ax.set_ylabel('Varış Gecikmesi (dk)')
plt.tight_layout()
plt.show()
```

```python
avg_delay = (
    df.groupby('Reporting_Airline')['ArrDelay']
    .mean()
    .sort_values(ascending=False)
    .head(10)
)

fig, ax = plt.subplots(figsize=(9, 4))
avg_delay.plot(kind='bar', ax=ax, edgecolor='white')
ax.set_title('Havayoluna Göre Ortalama Varış Gecikmesi — Top 10', fontsize=13)
ax.set_xlabel('Havayolu')
ax.set_ylabel('Ort. ArrDelay (dk)')
ax.tick_params(axis='x', rotation=0)
plt.tight_layout()
plt.show()
```

## Bölüm 6 — Özet Bulgular

### Kritik Bulgular

**Bölüm 1 — Genel Bakış**
- Beklenen boyut: ~500k–700k satır, ~110 kolon.
- `object` tipinde gelen sayısal sütunlar transform adımında cast edilmeli.
- `CRSDepTime` / `CRSArrTime` `0–2359` integer formatında — pipeline'da `//100` ile saat, `%100` ile dakika çıkarılmalı.

**Bölüm 2 — Null Analizi**
- Delay kolonlarındaki null'lar beklenen: iptal uçuşlarda gecikme verisi oluşmuyor → transform'da bu null'ları olduğu gibi koru.
- `CancellationCode` iptal olmayan satırlarda null → staging'de bu kolon `nullable` olmalı.
- `DepTime` / `ArrTime` iptal uçuşlarda null → sentinel değerle doldurmak yerine null bırak.

**Bölüm 3 — Dağılımlar**
- `Year` ve `Month` sabitse partition metadata olarak tutulabilir, kolon olarak taşınmayabilir.
- `Distance` sağa çarpık — min değer kontrol edilmeli, 0 veya negatif olmamalı.

**Bölüm 4 — İptal**
- İptal oranı genellikle %1–3. Ocak kış ayı olduğu için Weather (B) kodunun ağırlıklı olması beklenir.
- Havayoluna göre normalize edilmiş oran SLA takibinde kullanılabilir.

**Bölüm 5 — Gecikme**
- `is_delayed` = `ArrDelay > 15` → pipeline'da boolean kolon olarak türetilecek.
- `DepDelay` ile `ArrDelay` yüksek korelasyon — regresyon modeli için güçlü feature.
- `LateAircraftDelay` genellikle en yüksek paya sahip neden — zincir gecikme etkisi.

---

### Transform Adımı İçin Yapılacaklar (`transform_load_staging.py`)

1. **Tip dönüşümleri:** `object` kolonları `pd.to_numeric(errors='coerce')` ile cast et; başarısız olanları logla.
2. **Zaman dönüşümü:** `CRSDepTime`, `CRSArrTime`, `DepTime`, `ArrTime` → HH:MM veya `TimestampType`.
3. **Türetilmiş kolonlar:** `is_delayed` (ArrDelay > 15), `delay_category`, `taxi_time` (ActualElapsedTime − AirTime).
4. **Null stratejisi:** Delay kolonları iptal uçuşlarda null kalmalı — 0 ile doldurma.
5. **Aykırı değer kontrolü:** `Distance <= 0`, `ArrDelay < -120` veya `> 600` gibi değerleri logla.
6. **Partition:** `FlightDate` ile Parquet partition → okuma hızı artar.
7. **Sütun seçimi:** ~110 kolonun tamamı staging'e gitmek zorunda değil — business-critical ~30–40 kolon belirlenmeli.
